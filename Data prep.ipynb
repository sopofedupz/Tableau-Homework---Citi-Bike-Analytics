```python
# import dependencies
import pandas as pd
import datetime as dt
```

```python
# read datasets
mpls_Maydata = pd.read_csv("data/201905-niceride-tripdata.csv")
mpls_Junedata = pd.read_csv("data/2001906-niceride-tripdata.csv")
mpls_Julydata = pd.read_csv("data/201907-niceride-tripdata.csv")
mpls_data = mpls_Maydata.append(mpls_Junedata)
mpls_data = mpls_data.append(mpls_Julydata)

nyc_Maydata = pd.read_csv("data/JC-201905-citibike-tripdata.csv")
nyc_Junedata = pd.read_csv("data/JC-201906-citibike-tripdata.csv")
nyc_Julydata = pd.read_csv("data/JC-201907-citibike-tripdata.csv")
nyc_data = nyc_Maydata.append(nyc_Junedata)
nyc_data = nyc_data.append(nyc_Julydata)

# rename column
nyc_data = nyc_data.rename(columns = {'stoptime': 'end_time', 'starttime': 'start_time'})

# add data source
mpls_data['city'] = "Minneapolis"
nyc_data['city'] = "NYC"

# # merge the two dataset
bike_data = mpls_data.append(nyc_data, sort = False)
```
